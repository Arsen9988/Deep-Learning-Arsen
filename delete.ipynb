```python







# import tensorflow as tf

# # Anvnd alla tillgngliga CPU-krnor
# tf.config.threading.set_intra_op_parallelism_threads(0)  # Lt TensorFlow bestmma optimalt antal
# tf.config.threading.set_inter_op_parallelism_threads(0)

# # Bekrfta att endast CPU anvnds
# tf.debugging.set_log_device_placement(True)  # Loggar vilka enheter som anvnds






import keras  # Keras r ett API fr att bygga och trna neurala ntverk, som ofta anvnds med TensorFlow som backend.
from keras import layers  # layers anvnds fr att bygga de olika delarna (lager) i ett neuralt ntverk, t.ex. Dense- och Convolutional-lager.
import gymnasium as gym  # Gymnasium r en plattform fr att skapa miljer fr reinforcement learning (RL).
from gymnasium.wrappers.frame_stack import FrameStack  
# FrameStack r en wrapper frn Gymnasium som kombinerar flera p varandra fljande frames till en enda observation.
# Detta r anvndbart i spel dr kontext frn tidigare frames behvs, exempelvis rrelseriktning i Atari-spel.

from gymnasium.wrappers.atari_preprocessing import AtariPreprocessing  
# AtariPreprocessing r en inbyggd funktion fr att frbehandla bilder frn Atari-spel.
# Den konverterar bilder till grskala, minskar deras storlek och beskr dem fr att underltta inlrning.

import numpy as np  # NumPy anvnds fr numeriska operationer som matriser, slumpval och matematiska berkningar.
import tensorflow as tf  # TensorFlow anvnds som backend fr att implementera och trna neurala ntverk.
import ale_py  # ale_py r en wrapper fr Arcade Learning Environment (ALE) som tillhandahller Atari-spel till Gymnasium.











# Hyperparametrar fr RL-trningen
seed = 42  # Slumpfr fr att gra experimenten reproducerbara.
gamma = 0.99  # Diskonteringsfaktor fr framtida belningar. Vrden nra 1 prioriterar lngsiktiga belningar.
epsilon = 1.0  # Startvrde fr epsilon i epsilon-greedy-algoritmen, som anvnds fr att utforska.
epsilon_min = 0.1  # Minsta vrde fr epsilon, vilket begrnsar slumpmssiga val under utforskning.
epsilon_max = 1.0  # Maxvrde fr epsilon, anvnds vid start.
epsilon_interval = (epsilon_max - epsilon_min)  
# Intervall fr att gradvis minska epsilon under trning, frn maximal utforskning till kontrollerad utforskning.

batch_size = 32  # Antal exempel som anvnds vid varje trningssteg.
max_steps_per_episode = 10000  # Max antal steg i en episod innan den avslutas.
max_episodes = 0  # Max antal episoder (0 innebr obegrnsat).
max_frames = 1e6  # Max antal bildramar som ska bearbetas innan trningen avslutas.







# Registrerar Atari-miljer via ALE.
gym.register_envs(ale_py)
# Skapar Atari Breakout-miljn.
env = gym.make("SpaceInvadersNoFrameskip-v4", render_mode="rgb_array")  
# Spelmilj: Breakout utan att hoppa ver ngra frames. render_mode="rgb_array" gr att vi fr bilddata.
env = AtariPreprocessing(env)  

# Applicerar frbehandling p miljn. Detta inkluderar:
# - Omvandling till grskala
# - Bildskalning till 84x84 pixlar
# - Normalisering av pixlar.

env = FrameStack(env, 4)  
# Staplar fyra p varandra fljande frames fr att ge modellen temporal kontext (rrelseinformation).

# Triggerfunktion fr att spela in video var 1000:e timsteg.
trigger = lambda t: t % 500 == 0
env = gym.wrappers.RecordVideo(env, video_folder="videos", episode_trigger=trigger, disable_logger=True)  
# Wrapper fr att spela in videos av episoder som uppfyller trigger-funktionen.

num_actions = env.action_space.n
  # Antal mjliga handlingar i Breakout (t.ex. vnster, hger, stanna, skjut).
  
action_meanings = env.unwrapped.get_action_meanings()
print(f"{num_actions} {action_meanings}") # 'NOOP', 'FIRE', 'RIGHT', 'LEFT', 'RIGHTFIRE', 'LEFTFIRE'








def create_q_model():
    return keras.Sequential(
        [
            layers.Conv2D(32, kernel_size=8, strides=4, activation="relu"),  
            # Frsta konvolutionslagret. 
            # - 32 filter, storlek 8x8, med steg 4.
            # - Aktiveringsfunktion: ReLU fr att introducera icke-linearitet.
            
            layers.Conv2D(64, kernel_size=4, strides=2, activation="relu"),  
            # Andra konvolutionslagret.
            # - 64 filter, storlek 4x4, med steg 2.
            
            layers.Conv2D(64, kernel_size=3, strides=1, activation="relu"),  
            # Tredje konvolutionslagret.
            # - 64 filter, storlek 3x3, med steg 1.
            
            layers.Flatten(),  
            # Flatten omvandlar 2D/3D-utgng till en 1D-vektor fr att frbereda fr Dense-lager.

            layers.Dense(512, activation="relu"),  
            # Tckt lager med 512 neuroner. ReLU anvnds fr icke-linearitet.
            
            layers.Dense(num_actions, activation="linear")  
            # Utgngslager med antal neuroner som matchar antalet tgrder (num_actions). 
            # Aktiveringsfunktion: linear eftersom vi approximerar Q-vrden.
        ]
    )


def preprocess_input(data):
    return np.transpose(data, (1, 2, 0))








# Skapa den primra modellen och en target-modell (fr stabilitet vid uppdateringar).
model = create_q_model()
model_target = create_q_model()

# Optimizer fr trning: Adam med en lrhastighet p 0.00025 och gradientklippning fr stabilitet.
optimizer = keras.optimizers.Adam(learning_rate=0.00025, clipnorm=1.0)

# Buffertar och andra variabler fr att lagra trningens historia.
action_history = []  # Historik av valda tgrder.
state_history = []  # Historik av observerade tillstnd.
state_next_history = []  # Historik av efterfljande tillstnd.
rewards_history = []  # Historik av erhllna belningar.
done_history = []  # Historik av episod-slutstatus.
episode_reward_history = []  # Belningar fr varje avslutad episod.
running_reward = 0  # Medelbelning ver senaste episoderna.
episode_count = 0  # Antal episoder hittills.
frame_count = 0  # Totalt antal processade frames.

# Parametrar fr epsilon-greedy policy.
epsilon_random_frames = 50000  # Antal frames dr endast slumpmssiga tgrder anvnds.
epsilon_greedy_frames = 1000000.0  # Antal frames ver vilka epsilon avtar frn 1.0 till 0.1.

# Maximal replay buffer-storlek och trningskonfiguration.
max_memory_length = 1000000  # Max antal lagrade vergngar i replay-buffer.
update_after_actions = 4  # Antal actions mellan varje trningsuppdatering.
update_target_network = 10000  # Frekvens (i frames) fr att uppdatera target-modellen.
loss_function = keras.losses.Huber()  # Huberfrlust, robust fr utliggare.






while True:  # En ondlig loop som trnar modellen tills ett avslutande villkor ns.
    observation, _ = env.reset()  # Nollstller miljn och hmtar den initiala observationen.
    state = np.array(observation)  # Konverterar observationen till en NumPy-array.
    episode_reward = 0  # Initierar episodens totala belning.

    for timestep in range(1, max_steps_per_episode):  # Begrnsar antalet steg per episod.
        frame_count += 1  # kar den totala bildramrknaren (anvnds fr tidsstyrda tgrder).

        # Epsilon-greedy policy: Vlj antingen en slumpmssig tgrd eller den bsta tgrden frn modellen.
        if frame_count < epsilon_random_frames or epsilon > np.random.rand(1)[0]:
            # Utforskar genom att ta en slumpmssig tgrd.
            action = np.random.choice(num_actions)
        else:
            # Uppdatera epsilon-greedy steget
            state_processed = preprocess_input(state) #------------------------------------------------
            state_tensor = keras.ops.convert_to_tensor(state_processed)
            state_tensor = keras.ops.expand_dims(state_tensor, 0)
            action_probs = model(state_tensor, training=False)
            # # Exploaterar genom att anvnda modellen fr att vlja den bsta tgrden.
            # state_tensor = keras.ops.convert_to_tensor(state)  # Konverterar tillstndet till en tensor.
            # state_tensor = keras.ops.expand_dims(state_tensor, 0)  # Lgger till batch-dimension.
            # action_probs = model(state_tensor, training=False)  # Frutspr Q-vrden utan att aktivera trningslge.
            # action = keras.ops.argmax(action_probs[0]).numpy()  # Vljer tgrden med hgst Q-vrde.

        # Minskar epsilon linjrt fr att gradvis minska sannolikheten fr slumpmssiga tgrder.
        epsilon -= epsilon_interval / epsilon_greedy_frames
        epsilon = max(epsilon, epsilon_min)  # Begrnsar epsilon till minimumvrdet.

        # Utfr tgrden i miljn och observerar resultatet.
        state_next, reward, done, _, _ = env.step(action)  
        state_next = np.array(state_next)  # Konverterar nsta tillstnd till NumPy-array.

        episode_reward += reward  # Lgg till belningen frn detta steg till episodens totala belning.

        # Spara vergngen i replay-buffer.
        action_history.append(action)
        state_history.append(state)
        state_next_history.append(state_next)
        done_history.append(done)
        rewards_history.append(reward)

        # G vidare till nsta tillstnd.
        state = state_next
        
        # Trna modellen efter var fjrde tgrd och om replay-buffer r tillrckligt stor.
        if frame_count % update_after_actions == 0 and len(done_history) > batch_size:
            # Vljer slumpmssigt ett urval av tidigare vergngar fr att skapa trningsdata.
            indices = np.random.choice(range(len(done_history)), size=batch_size)

            # Extraherar de valda vergngarna frn replay-buffer.
            state_sample = np.array([state_history[i] for i in indices])
            state_next_sample = np.array([state_next_history[i] for i in indices])
            rewards_sample = [rewards_history[i] for i in indices]
            action_sample = [action_history[i] for i in indices]
            done_sample = keras.ops.convert_to_tensor(
                [float(done_history[i]) for i in indices]
            )

            state_next_sample_processed = np.array([preprocess_input(s) for s in state_next_sample])
            future_rewards = model_target.predict(state_next_sample_processed, verbose=0)

            #future_rewards = model_target.predict(state_next_sample, verbose=0)---------------------------------------------------------
            
            
            
            
            
            
            # Uppdaterade Q-vrden: belning + gamma * max framtida belning.
            updated_q_values = rewards_sample + gamma * keras.ops.amax(future_rewards, axis=1)

            # Stter Q-vrdet till -1 om tillstndet r terminalt.
            updated_q_values = updated_q_values * (1 - done_sample) - done_sample

            # Skapar en mask fr att berkna frlust endast fr den utfrda tgrden.
            masks = keras.ops.one_hot(action_sample, num_actions)

            with tf.GradientTape() as tape:
                # Frutspr Q-vrden fr det aktuella tillstndet.
                #q_values = model(state_sample)------------------------------------------------------------------------------------------------
            
            

                state_sample_processed = np.array([preprocess_input(s) for s in state_sample])
                q_values = model(state_sample_processed)


                # Extraherar Q-vrdet fr den valda tgrden.
                q_action = keras.ops.sum(keras.ops.multiply(q_values, masks), axis=1)

                # Berknar frlusten mellan de frvntade och de faktiska Q-vrdena.
                loss = loss_function(updated_q_values, q_action)

            # Utfr backpropagation fr att uppdatera modellens viktparametrar.
            grads = tape.gradient(loss, model.trainable_variables)
            optimizer.apply_gradients(zip(grads, model.trainable_variables))

        # Uppdaterar target-modellen efter ett fast antal frames.
        if frame_count % update_target_network == 0:
            model_target.set_weights(model.get_weights())  # Kopierar vikterna frn trningsmodellen.
            print(f"Best score of last 100: {np.max(episode_reward_history)}, running reward: {running_reward:.2f} at episode {episode_count}, frame count {frame_count}")
            model.save(f"models/breakout_qmodel_{episode_count}.keras")  # Sparar modellen efter varje uppdatering.

        # Begrnsar replay-buffer till maxlngd genom att ta bort gamla vergngar.
        if len(rewards_history) > max_memory_length:
            del rewards_history[:1]
            del state_history[:1]
            del state_next_history[:1]
            del action_history[:1]
            del done_history[:1]

        # Om episoden r klar, avsluta loopen fr denna episod.
        if done:
            break

    # Uppdaterar lpande belning baserat p de senaste 100 episoderna.
    episode_reward_history.append(episode_reward)
    if len(episode_reward_history) > 100:
        del episode_reward_history[:1]
    running_reward = np.mean(episode_reward_history)

    # kar episodrknaren.
    episode_count += 1
    print(f"Episode {episode_count-1}: {episode_reward}")

    # Kontrollera om problemet anses lst baserat p en belningstrskel.
    if running_reward > 800:
        print("Solved at episode {}!".format(episode_count))
        break

    # Avbryter om max episoder eller max frames har ntts.
    if max_episodes > 0 and episode_count >= max_episodes:
        print("Stopped at episode {}!".format(episode_count))
        break
    if max_frames <= frame_count:
        print(f"Stopped at frame {frame_count}!")
        break
```
